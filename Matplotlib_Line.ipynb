# Linestyle
You can use the keyword argument linestyle, or shorter ls, to change the style of the plotted line:

# Shorter Syntax
The line style can be written in a shorter syntax:

linestyle can be written as ls.

dotted can be written as :.

dashed can be written as --.

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x_axis = np.array([1,2,6,3,7])
y_axis = np.array([7,2,4,1,5])
```

```python
plt.plot(x_axis,y_axis,ls = 'dotted')
plt.show()
```

```python
plt.plot(x_axis,y_axis,ls = ':')
plt.show()
```

```python
plt.plot(x_axis,y_axis,ls = 'dashdot')
plt.show()
```

```python
plt.plot(x_axis,y_axis,ls = 'dashdot',color = 'g')
plt.show()
```

# Line Width
You can use the keyword argument linewidth or the shorter lw to change the width of the line.

```python
plt.plot(x_axis,y_axis,ls = 'dotted',lw = '10.5')
plt.show()
```

```python
plt.plot(x_axis,y_axis,y_axis,x_axis)
plt.show()
```
